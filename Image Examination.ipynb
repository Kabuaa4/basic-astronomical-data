# Image Examination

Let us play around with some very quick visualizations related to images. We have already seen how to use ds9 for looking at images but let us now connect up our image viewer with our code and actually interactively explore the image properties in different regions.

Once upon a time, we would use a software called IRAF (Image Reduction and Analysis Facility) in order to do this but now this project is being retired and people are being encouraged to find alternatives. So, let's see if we can use Python for this. In principle, it should be possible to make Python interact with ds9 but this requires setting up a software called XPA to enable communication between ds9 and Python. Instead we will use ginga to help make things a little easier

```python
import imexam
%matplotlib
```

```python
viewer = imexam.connect(viewer='ginga')
```

You should see a new tab pop-up which essentially is the Ginga viewer HTML5 widget. It might make sense to separate this tab into a window and reduce its size to have both the current notebook and the image viewer side by side.

```python
viewer.load_fits('example.fits')
```

Now, we will enter the interactive mode. What to do after this? Go back to the viewer window, hit 'i'. And then you can point your cursor/pointer to various regions and try out various commands.

```python
viewer.imexam()
```

```python
viewer.close()
```
